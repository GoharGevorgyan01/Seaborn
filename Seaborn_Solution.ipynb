```python
# НАПИШИТЕ ВАШ КОД В ЭТОЙ ЯЧЕЙКЕ
employed = df[df['DAYS_EMPLOYED'] <0]
employed['DAYS_EMPLOYED']=employed['DAYS_EMPLOYED']* -1
employed['DAYS_BIRTH']=employed['DAYS_BIRTH']* -1
```

```python
employed[['DAYS_BIRTH','DAYS_EMPLOYED']]
```

```python
plt.figure(figsize = (11,5),dpi=160)
sns.scatterplot(x='DAYS_BIRTH',y='DAYS_EMPLOYED',data=employed,alpha =0.008,lw = 0)
          
```

Графики boxplot отображены для каждого значения колонки NAME_FAMILY_STATUS, показывая распределения их общего дохода. Раскраска hue сделана по колонке "FLAG_OWN_REALTY". ### ЗАДАНИЕ: воссоздайте приведённый ниже график Heat Map:

<img src='heatmap_solution.png'>

**Эта тепловая карта heatmap показывает корреляцию между колонками в наборе данных. Корреляцию можно посчитать с помощью метода .corr().  Обратите внимание, что колонка FLAG_MOBIL имеет NaN-корреляцию со всеми другими колонками - это видно в результатах метода .corr(). Поэтому Вам следует удалить колонку FLAG_MOBIL перед построением графика.**

```python
df2 = df.drop('FLAG_MOBIL',axis=1).corr()
df2
```

```python
# НАПИШИТЕ ВАШ КОД В ЭТОЙ ЯЧЕЙКЕ
sns.heatmap(data = df2,cmap = 'viridis')
```
